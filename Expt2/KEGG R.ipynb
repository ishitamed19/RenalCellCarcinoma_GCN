```r
library(KEGGgraph)
```

```r
## LOOP Through all xml files and generate its csv
```

```r
files <- list.files(path="../../Data/KEGG", pattern="*.xml", full.names=TRUE, recursive=FALSE)
```

```r
lapply(files, function(x) {
    gdf <- parseKGML2DataFrame(x,expandGenes=TRUE)
    print(substr(x,17,24))
    p = file.path("../../Data/KEGG_csv",(paste(substr(x,17,24), ".csv")))
    write.csv(gdf,p, row.names = TRUE)
})
```
